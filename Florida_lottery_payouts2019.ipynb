```python
#Utilizing this function written by Rafael Valero-Fernandez to write the dataframe to MongoDB
def write_df_to_mongoDB(  my_df,\
                          database_name = 'mydatabasename' ,\
                          collection_name = 'mycollectionname',
                          server = 'localhost',\
                          mongodb_port = 27017,\
                          chunk_size = 100):
    #"""
    #This function take a list and create a collection in MongoDB (you should
    #provide the database name, collection, port to connect to the remoete database,
    #server of the remote database, local port to tunnel to the other machine)
    #
    #---------------------------------------------------------------------------
    #Parameters / Input
    #    my_list: the list to send to MongoDB
    #    database_name:  database name
    #
    #    collection_name: collection name (to create)
    #    server: the server of where the MongoDB database is hosted
    #        Example: server = '132.434.63.86'
    #    this_machine_port: local machine port.
    #        For example: this_machine_port = '27017'
    #    remote_port: the port where the database is operating
    #        For example: remote_port = '27017'
    #    chunk_size: The number of items of the list that will be send at the
    #        some time to the database. Default is 100.
    #
    #Output
    #    When finished will print "Done"
    #----------------------------------------------------------------------------
    #FUTURE modifications.
    #1. Write to SQL
    #2. Write to csv
    #----------------------------------------------------------------------------
    #30/11/2017: Rafael Valero-Fernandez. Documentation
    #17/11/2019: Martie Feldtmann - updated the range to a list for Python 3
    #"""



    #To connect
    # import os
    # import pandas as pd
    # import pymongo
    # from pymongo import MongoClient

    client = MongoClient('localhost',int(mongodb_port))
    db = client[database_name]
    collection = db[collection_name]
    # To write
    collection.delete_many({})  # Destroy the collection
    #aux_df=aux_df.drop_duplicates(subset=None, keep='last') # To avoid repetitions
    my_list = my_df.to_dict('records')
    l =  len(my_list)
    ran = list(range(l)) # MF (11/11/2019) needed to change the range to list to append
    steps=ran[chunk_size::chunk_size]
    steps.extend([l])

    # Inser chunks of the dataframe
    i = 0
    for j in steps:
        print(j)
        collection.insert_many(my_list[i:j]) # fill de collection
        i = j

    print('Done')
    return
```

### Use regular expressions to create patterns for extracting certain parts of the raw data

```python
import re
from datetime import datetime
import time 

#Use regular expression to extract the dates from the raw data
regex = re.compile(r'DATE:\s\w{3,12}\s\w{1,2}\,\s\w{4}')
Dates = regex.findall(raw)
#Remove the string Date: from the list
Dates = [item.replace('DATE: ','') for item in Dates]
#change the date format
Dates = [datetime.strptime(str(i), "%B %d, %Y").strftime("%m/%d/%Y") for i in Dates]
print(Dates[0:5])
print(len(Dates))
```

```python
#Using regular expression to extract the winning numbers from the raw text
regex = re.compile(r'WINNING NUMBERS:\s\s\w{2}\-\w{2}\-\w{2}\-\w{2}\-\w{2}')
winning_numbers = regex.findall(raw)
#Remove the string WINNING NUMBERS: from the list
winning_numbers = [item.replace('WINNING NUMBERS: ','') for item in winning_numbers]
print(winning_numbers[0:5])
print(len(winning_numbers))
```

```python
#Use regulr expression to find the multiplier
regex = re.compile(r'X\d')
multiplier = regex.findall(raw)
print(multiplier[0:5])
print(len(multiplier))
```

```python
#Use regulr expression to find the total sales
#regex = re.compile(r'Total Sales:\s\$\s\d\,\d{3}\,\d{3}\.\d{2}')
regex = re.compile(r'Total Sales:\s\$\s*[0-9]+[0-9,.]*[0-9]*')
total_sales = regex.findall(raw)
#Remove the string Total Sales: from the list
total_sales = [item.replace('Total Sales: ','') for item in total_sales]
#Remove the space after the $ from the list
total_sales = [item.replace(' ','') for item in total_sales]
print(total_sales[0:5])
print(len(total_sales))
```

```python
#Use regulr expression to find the total winners
regex = re.compile(r'Total Winners:\s\d{2}\,\d{3}')
total_winners = regex.findall(raw)
#Remove the string Total Winners: from the list
total_winners = [item.replace('Total Winners: ','') for item in total_winners]
print(total_winners[0:5])
print(len(total_winners))
```

```python
import numpy as np
combined_list_lotto = pd.DataFrame(np.column_stack([Dates, winning_numbers, multiplier, total_sales, total_winners]), 
                               columns=['Date', 'Winning Numbers', 'Multiplier', 'Total Sales', 'Total Winners'])


#Split the numbers into separate columns

combined_list_lotto=combined_list_lotto.join(combined_list_lotto['Winning Numbers'].str.split('-', expand=True)) #it is dropping the last number
#rename the new columns
combined_list_lotto=combined_list_lotto.rename(columns = {0:'Ball 1',1:'Ball 2', 2:'Ball 3',3:'Ball 4', 4:'Ball 5'})
#change the order of the columns
combined_list_lotto = combined_list_lotto[['Date', 'Total Sales','Total Winners','Ball 1','Ball 2', 'Ball 3', 'Ball 4', 'Ball 5', 'Multiplier']]
#remove the $ sign and convert to float
combined_list_lotto['Total Sales']=combined_list_lotto['Total Sales'].str.replace(',', '', regex=False)
combined_list_lotto['Total Sales'] = [x.strip('$') for x in combined_list_lotto['Total Sales']]
combined_list_lotto['Total Sales'] = pd.to_numeric(combined_list_lotto['Total Sales']) 
combined_list_lotto['Total Sales'] = [round(x,2) for x in combined_list_lotto['Total Sales']]
combined_list_lotto.head()
```

```python
#Call the write to MongoDB function to add the dataframe to the db
write_df_to_mongoDB(  combined_list_lotto,\
                          database_name = 'Florida_Lottery' ,\
                          collection_name = 'lotto')
```
