```python
#Dependencies
import re
import pandas as pd
from datetime import datetime
```

```python
#URL of page to be scrapes
link= 'https://lotterycorner.com/ny/powerball/2019'
```

```python
tables=pd.read_html(link)
```

```python
org_df=tables[0]
org_df.columns= ['Date', 'Winning Numbers', 'Jackpot']
del org_df['Winning Numbers']
org_df.head()

org_df=org_df.iloc[2:]
org_df.head()

org_df
```

```python
from datetime import datetime

#dr=org_df.Date
#date_d1 = [datetime.strptime(dr,"%B %d, %Y").strftime("%m/%d/%Y")]
#print(date_d1)
#int_dates= int(date_d1)
#print(int_dates)
#dr=org_df.Date
#org_df['Date']=[datetime.strptime(dr,"%B %d, %Y").strftime("%m/%d/%Y")] 

#[datetime.strptime(dr,"%B %d, %Y").strftime("%m/%d/%Y")](lambda x: datetime.strptime(x, '%Y%m%d%H'))

date=list(map(lambda x: datetime.strptime(x,'%b %d, %Y').strftime('%m/%d/%Y'), org_df['Date']))
date
#pd.to_datetime[(org_df['Date'], format='%b %d, %Y').strftime("%m/%d/%Y")]




```

```python
#make sure only millions 
mask=org_df['Jackpot'].str.contains('thousand')
org_df[mask]

pattern= re.compile(r'$/d+')
re.findall(pattern,str(org_df['Jackpot']))
```

```python
values=[]

for i in org_df['Jackpot'].str.split():
    values.append(i[0][1:])



    
    
    
```

```python
values= pd.Series(values)
values=values.astype(int)
values=values * 1000000
values
```

```python
import numpy as np
combined_list_pb = pd.DataFrame(np.column_stack([date, values]),
                               columns=['Date', 'Jackpot'])
```

```python
combined_list_pb
```

```python
#To connect
import os
import pymongo
from pymongo import MongoClient



# Create connection variable
conn = 'mongodb://localhost:27017'
pb_db= 'NewYork_Lottery'
pb_cl= 'Powerball'

# Pass connection to the pymongo instance.
client = pymongo.MongoClient(conn)

# Connect to a database. Will create one if not already available.
db = client[pb_db]
cl = db[pb_cl]

# Drops collection if available to remove duplicates
db.cl.drop()
cleandf=combined_list_pb.set_index('Date')
writedf=cleandf.to_dict('index')

cl.insert(writedf)


```
